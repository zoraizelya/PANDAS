```python
t3 = titanic[["age in days","fare"]]
t3.head()
```

# Method of Normalization 
1. Simple feature scaling 
    - x(new)=x(old) / x(max)
2. Min / Max method
3. Z-score (standard score) -3 to +3
4. Log Transformation

```python
# simple feature scaling
t3['fare']=t3['fare']/t3['fare'].max()
t3['age in days']=t3['age in days']/t3['age in days'].max()
t3.head(5)
```

```python
# Min / Max method
t3['fare']=(t3['fare']-t3['fare'].min()) / (t3['fare'].max()-t3['fare'].min())
t3.head()
```

```python
# Z-score 
t3['fare']=(t3['fare']-t3['fare'].mean()) / t3['fare'].std()
t3.head()
```

# Binning
- grouping the values into smaller number of values (bins)
- convert numeric into categories (baby, adult , old) or 1-10 , 11-30 , 31-100 etc
- to have better understanding of groups
    - low vs mid vs high 

```python
bins= np.linspace(min(tnc['age']),max(tnc['age']),15000) 
# age_groups=["Baby","Adults","Old"]
tnc['age'] = pd.cut(tnc['age'],bins=[0,10,40,160],labels=["Baby","Adults","Old"],include_lowest=True)
tnc['age'].head()
```
